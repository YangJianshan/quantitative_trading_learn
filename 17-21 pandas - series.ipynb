## 19 series数据对齐

```python
sr1 = pd.Series([12, 23, 34], index=['c', 'a', 'd'])
sr2 = pd.Series([11, 20, 10], index=['d', 'c', 'a'])
sr1 + sr2
```

```python
sr2['b'] = 12
sr2
```

```python
sr1+sr2
```

```python
sr1.add(sr2,fill_value=0)
```

## 20 series缺失数据处理

```python
# 删除缺失数据
sr = sr1+sr2
sr
sr[sr.isnull()].values

# dropna 丢掉缺失值
sr.dropna()

# fillna 将nan的值赋值为指定值
sr = sr.fillna(sr.mean())
sr
```
