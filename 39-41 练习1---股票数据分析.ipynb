### 假如从2018年1月1日开始，每月第一个交易日买入1手股票，每年最后一个交易日卖出所有股票，到今天为止，收益如何

```python
# 找到2018.1.1每个月第一个交易日的记录，1.找到18年交易记录；2.找到每个月的第一个交易日的标签作为索引
import pandas as pd
df_300142 = pd.read_csv('沃森生物300142.csv', index_col='date', parse_dates=True)
df = df_300142.loc[:'2018',:]
df_monthly = df.resample('MS').first()  # 2018年之后每个月第一个交易日的数据
df_yearly = df.resample('A').last()  # 2018年之后每年最后一个交易日的数据
df_yearly = df_yearly[:-1]  # 将2020年切掉，因为还没到时间

# 每月第一个交易日 以 开盘价 买入1手
cost_money = 0  # 买入的花费
hold = 0  # 持有股票的总分数

# 1.先计算有几年的数据
for year in df_yearly.index:
    # 2. 将年份转换为字符串，用于找到本年的每月数据
    # 3. 以每月第一个交易日的开盘价格买入1手
    y = year.strftime('%Y')
    cost_money += df_monthly[y]['open'].sum()*100  # 买入花费
    hold += len(df_monthly[y]['open'])*100  # 持有份额
    
    # 4 每年的最后一个交易日，以收盘价卖出
    cost_money -= df_yearly[y]['close'].values[0]*hold  # 买入的花费 - 卖出的收益 = 现收益
    hold = 0
    pass

# 5 最后计算2020年还没有到年底卖出的花费和分数

cost_money



    
```
