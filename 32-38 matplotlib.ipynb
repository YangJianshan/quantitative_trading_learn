### 35 使用matplotlib绘制数学函数图像

### matplotlib实例：绘制数学函数图像
使用matplotlib模块在窗口中绘制数学函数y=x， y=x², y=3x²， y=3x³+5x²+2x+1
使用不同颜色的线加以区别，并使用图例说明个线代表的函数

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-100,100,10000)

y1 = x
plt.plot(x,y1, '-r', label='y=x')

y2 = x**2
plt.plot(x, y2, 'g', label='y=x²')

y3 = 3*(x**3) + 5*(x**2) + 2*x+1
plt.plot(x, y3, 'b', label='y=3x³+5x²+2x+1')

plt.legend()

plt.ylim(-100,100)
plt.xlim(-100,100)

plt.show()
```
