```python
# 已知若干家公司的市值（美元），将其转换为人民币
import numpy as np
import random
a = [random.uniform(100, 200) for i in range(10)]  # 生成10家公司的市值
x = 6.8  # 汇率
a = np.array(a)
print(a)
a*x
```

## 13numpy 布尔型索引

```python
a = [random.randint(0, 10) for i in range(100)]
a = np.array(a)
a>5
a[a>5]
```

```python
# 给一个数组，选取数组中所有大于5的偶数
a = [random.randint(0, 100) for i in range(100)]
a = np.array(a)
b = a[a>50 & (a%2==0)]
b[b%2==0]
```
